```python
#install the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#read the dataset
df=pd.read_csv("StressLevelDataset.csv")
df.head()
```

```python
#Understand the data
df.info()
df.isnull().sum().sort_values(ascending=False).head(10)
df.describe(include="all").T
```

```python
df.duplicated().sum()
```

```python
#Univariate Analysis on anxiety levels
sns.histplot(df['anxiety_level'],bins=20,kde=True)
plt.title('Distribution on Anxiety Levels')
plt.show()
```

```python
#Bivariate analysis

sns.scatterplot(x='sleep_quality', y='stress_level', data=df, alpha=0.5)
sns.regplot(x='sleep_quality', y='stress_level', data=df, scatter=False, color='red')
plt.title("Sleep Hours vs Stress Score")
plt.show()
```

```python
sns.lineplot(x='social_support',y='stress_level',data=df)
plt.title("Stress by social support level")
plt.show()
```

```python
(df.groupby('academic_performance')['stress_level'].mean().sort_values().plot(kind='bar',title='Avg stress by Academic Performance'))
plt.ylabel("stress_level")
plt.show()
```

```python
num_df = df.select_dtypes(include=np.number)
corr = num_df.corr(numeric_only=True)

# Mask correlations below absolute value of 0.3
mask = (corr.abs() < 0.3)

focus_cols = ['stress_level','anxiety_level','self_esteem','depression','sleep_quality','social_support']
sns.heatmap(
    df[focus_cols].corr(),
    cmap="Blues",
    annot=True,
    fmt=".2f",
    linewidths=0.5
)
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```
